Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
combined_data = pd.concat([baltimore_data, boston_data, new_york_data, philadelphia_data])
combined_data.head()
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.



```python
for col in combined_data.columns:
    pct_missing = np.mean(combined_data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Remove columns that have 100% missing values
cleaned_data = combined_data.drop(['Type', 'Grade', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode'], axis=1)
cleaned_data.head()
```

```python
#look for any columns that have all the same value
for col in cleaned_data.columns:
    print(f"{col} : unique items: {len(cleaned_data[col].unique())}")
```

```python
cleaned_data.shape
```

```python
#Drop columns that have only one unique value
cleaned_data = cleaned_data.drop(['Commodity Name'], axis=1)
cleaned_data.head()
```

```python
#Remove other columns that are not relevant
#Mostly Low: This column is not measured for pumpkins
#Mostly High: This column is not measured for pumpkins
cleaned_data = cleaned_data.drop(['Mostly Low', 'Mostly High'], axis=1)
cleaned_data.head()
```

```python
cleaned_data['Origin District'].unique()
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
Unit_of_Sale = cleaned_data.groupby(['Unit of Sale'])
Unit_of_Sale.first()
```

```python
mean_agg = Unit_of_Sale[['Low Price', 'High Price']].mean()
print(mean_agg)
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
Variety_per_market = cleaned_data.groupby(['City Name', 'Variety'])
Variety_per_market.first()
```
